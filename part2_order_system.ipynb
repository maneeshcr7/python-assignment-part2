```python
def add_item(cart, menu, item_name, quantity=1):
    """Adds an item to the cart or updates its quantity if it already exists."""
    # Check if the item exists in the menu
    if item_name not in menu:
        print(f"Error: '{item_name}' does not exist in the menu.")
        return cart
    
    # Check if the item is marked as available
    if not menu[item_name].get("available", False):
        print(f"Error: '{item_name}' is currently unavailable.")
        return cart

    # If the item is already in the cart, increase its quantity
    for entry in cart:
        if entry["item"] == item_name:
            entry["quantity"] += quantity
            return cart

    # Otherwise, add a new entry to the cart
    cart.append({
        "item": item_name,
        "quantity": quantity,
        "price": menu[item_name]["price"]
    })
    return cart


def remove_item(cart, item_name):
    """Removes an item from the cart completely."""
    for i, entry in enumerate(cart):
        if entry["item"] == item_name:
            del cart[i]
            return cart
            
    print(f"Notice: '{item_name}' is not in the cart.")
    return cart


def update_quantity(cart, item_name, new_quantity):
    """Updates the quantity of a specific item already in the cart."""
    for entry in cart:
        if entry["item"] == item_name:
            if new_quantity <= 0:
                # If quantity drops to 0 or below, remove the item
                return remove_item(cart, item_name)
            
            entry["quantity"] = new_quantity
            return cart
            
    print(f"Notice: '{item_name}' is not in the cart.")
    return cart


def print_order_summary(cart):
    """Calculates totals and prints the final formatted order summary."""
    print("\n========== Order Summary ==========")
    subtotal = 0.0
    
    for entry in cart:
        item_total = entry["price"] * entry["quantity"]
        subtotal += item_total
        # Left-align the item name to 18 chars, and quantity to 4 chars
        print(f"{entry['item']:<18} x{entry['quantity']:<4} ₹{item_total:.2f}")
        
    print("------------------------------------")
    
    gst = subtotal * 0.05
    total = subtotal + gst
    
    print(f"{'Subtotal:':<25}₹{subtotal:>7.2f}")
    print(f"{'GST (5%):':<25}₹{gst:>7.2f}")
    print(f"{'Total Payable:':<25}₹{total:>7.2f}")
    print("====================================")


# ==========================================
# SIMULATION SEQUENCE
# ==========================================

# 1. Define the mock menu and an empty cart
menu = {
    "Paneer Tikka": {"price": 180.0, "available": True},
    "Gulab Jamun": {"price": 60.0, "available": True},
    "Chicken Wings": {"price": 250.0, "available": False}
}

cart = []

print("--- Add 'Paneer Tikka' x 2 ---")
add_item(cart, menu, "Paneer Tikka", 2)
print("Cart State:", cart, "\n")

print("--- Add 'Gulab Jamun' x 1 ---")
add_item(cart, menu, "Gulab Jamun", 1)
print("Cart State:", cart, "\n")

print("--- Add 'Paneer Tikka' x 1 ---")
add_item(cart, menu, "Paneer Tikka", 1)
print("Cart State:", cart, "\n")

print("--- Try to add 'Mystery Burger' ---")
add_item(cart, menu, "Mystery Burger", 1)
print("Cart State:", cart, "\n")

print("--- Try to add 'Chicken Wings' ---")
add_item(cart, menu, "Chicken Wings", 1)
print("Cart State:", cart, "\n")

print("--- Remove 'Gulab Jamun' ---")
remove_item(cart, "Gulab Jamun")
print("Cart State:", cart, "\n")

# Print the final order summary
print_order_summary(cart)
```

```python
import copy

# 1. Setup Mock Data (Based on previous task context)
cart = [
    {"item": "Paneer Tikka", "quantity": 3, "price": 180.0}
]

inventory = {
    "Paneer Tikka": {"stock": 10, "reorder_level": 8},
    "Gulab Jamun": {"stock": 5, "reorder_level": 2},
    "Chicken Wings": {"stock": 0, "reorder_level": 5}
}

# 2. Deep Copy and Demonstration
# We use deepcopy because the inventory is a nested dictionary. 
# A shallow copy (inventory.copy()) would still share references to the inner dictionaries.
inventory_backup = copy.deepcopy(inventory)

print("--- Deep Copy Demonstration ---")
# Manually change a stock value to prove the backup is independent
inventory["Paneer Tikka"]["stock"] = 999
print(f"Modified Inventory (Paneer Tikka): {inventory['Paneer Tikka']['stock']}")
print(f"Backup Inventory (Paneer Tikka):   {inventory_backup['Paneer Tikka']['stock']}")

# Restore inventory to its original state
inventory = copy.deepcopy(inventory_backup)
print("Inventory successfully restored to original state.\n")

# 3. Simulate Order Fulfillment
print("--- Order Fulfillment ---")
for cart_item in cart:
    item_name = cart_item["item"]
    qty_requested = cart_item["quantity"]
    
    if item_name in inventory:
        available_stock = inventory[item_name]["stock"]
        
        # Deduct normally if stock is sufficient
        if available_stock >= qty_requested:
            inventory[item_name]["stock"] -= qty_requested
            print(f"Successfully deducted {qty_requested} unit(s) of '{item_name}'.")
        else:
            # Handle insufficient stock (deduct only what is available, don't drop below 0)
            print(f"Warning: Insufficient stock for '{item_name}'. Requested {qty_requested}, but only {available_stock} available.")
            inventory[item_name]["stock"] = 0
            print(f"Deducted {available_stock} unit(s) of '{item_name}'.")
    else:
        print(f"Warning: '{item_name}' does not exist in inventory.")

# 4. Reorder Alerts
print("\n--- Reorder Alerts ---")
for item_name, details in inventory.items():
    current_stock = details["stock"]
    reorder_level = details["reorder_level"]
    
    if current_stock <= reorder_level:
        print(f"⚠ Reorder Alert: {item_name} — Only {current_stock} unit(s) left (reorder level: {reorder_level})")

# 5. Final State Verification
print("\n--- Final State Verification ---")
print("Final Inventory:")
for item, details in inventory.items():
    print(f"  {item}: {details}")

print("\nBackup Inventory:")
for item, details in inventory_backup.items():
    print(f"  {item}: {details}")
```

```python
from collections import Counter

# 1. Setup Mock Initial Data
sales_log = {
    "2025-01-01": [
        {"order_id": 1, "items": ["Paneer Tikka", "Garlic Naan"], "total": 220.0},
        {"order_id": 2, "items": ["Gulab Jamun", "Veg Soup"], "total": 210.0}
    ],
    "2025-01-02": [
        {"order_id": 3, "items": ["Butter Chicken", "Garlic Naan"], "total": 350.0},
        {"order_id": 4, "items": ["Garlic Naan", "Veg Soup"], "total": 150.0}
    ]
}

def print_revenue_and_best_day(log: dict):
    """Calculates and prints revenue per day and the best-selling day."""
    revenue_per_day = {}
    
    for date, orders in log.items():
        daily_total = sum(order["total"] for order in orders)
        revenue_per_day[date] = daily_total
        
    print("--- Revenue Per Day ---")
    for date, total in revenue_per_day.items():
        print(f"{date}: ₹{total:.2f}")
        
    # max() with key allows us to efficiently find the dictionary key with the highest value
    best_day = max(revenue_per_day, key=revenue_per_day.get)
    best_total = revenue_per_day[best_day]
    print(f"\n🌟 Best-Selling Day: {best_day} with ₹{best_total:.2f}\n")

def print_most_ordered_item(log: dict):
    """Finds the item that appears in the greatest number of individual orders."""
    item_order_counts = Counter()
    
    for date, orders in log.items():
        for order in orders:
            # We use set() to ensure that if an item is listed twice in the SAME order,
            # it only counts as 1 order appearance.
            unique_items_in_order = set(order["items"])
            for item in unique_items_in_order:
                item_order_counts[item] += 1
                
    if item_order_counts:
        most_common_item, count = item_order_counts.most_common(1)[0]
        print(f"🏆 Most Ordered Item: {most_common_item} (appeared in {count} orders)\n")

def print_enumerated_order_list(log: dict):
    """Prints a formatted, numbered list of all orders using enumerate."""
    print("--- All Orders ---")
    
    # Flattening the nested dictionary into a list of tuples for easy enumeration
    all_orders = [(date, order) for date, orders in log.items() for order in orders]
    
    for index, (date, order) in enumerate(all_orders, start=1):
        items_str = ", ".join(order["items"])
        print(f"{index}.  [{date}] Order #{order['order_id']}  — ₹{order['total']:.2f} — Items: {items_str}")
    print()


# ==========================================
# EXECUTION SEQUENCE
# ==========================================

print("=== INITIAL SALES DATA ===")
print_revenue_and_best_day(sales_log)
print_most_ordered_item(sales_log)

# Add new data
print("=== ADDING NEW DAY (2025-01-05) ===\n")
sales_log["2025-01-05"] = [
    {"order_id": 11, "items": ["Butter Chicken", "Gulab Jamun", "Garlic Naan"], "total": 490.0},
    {"order_id": 12, "items": ["Paneer Tikka", "Rasgulla"],                     "total": 260.0},
]

print("=== UPDATED SALES DATA ===")
print_revenue_and_best_day(sales_log)

print_enumerated_order_list(sales_log)
```
